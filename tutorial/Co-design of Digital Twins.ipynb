### Create draft components
To create a draft component for your workflow, do the following:
1) Select a notebook code cell (e.g. the `# Data explorer` below) and open the NaaVRE containerizer:<br/>
<img src="./images/containerizer.png" alt="containerizer.png" width="600"/>
2) Press `Analyze cell`
3) Press `Create`:<br/>
<img src="./images/create.png" alt="create.png" width="300"/>

### Compose a workflow
To use the draft components to design a workflow, do the following:
1) Open a new launcher:<br/>
<img src="./images/new_launcher.png" alt="new_launcher.png" width="600"/> 
2) Open the experiment manager:<br/>
<img src="./images/experiment_manager.png" alt="experiment_manager.png" width="300"/>
3) Find the workflow component you have created in the catalog.
4) Drag and drop the component onto the canvas.<br/>
<img src="./images/component_to_canvas.png" alt="component_to_canvas.png" width="600"/>
<br/><br/>
Create another draft component from another jupyter notebook cell before continuing with the following steps.
<br/><br/>
6) Drag and drop the second component onto the canvas.
7) Connect the two blocks of the draft components on your canvas:<br/>
<img src="./images/connect_components.png" alt="connect_components.png" width="800"/><br/>
You have succesfully built a draft workflow.
<br/><br/>
Continue creating draft components and adding them to the workflow to design a workflow that you could use in your own research.
